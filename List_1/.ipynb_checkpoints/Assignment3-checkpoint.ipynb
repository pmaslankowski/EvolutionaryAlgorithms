```python
import time
import multiprocessing
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import mode
```

```python
def pbil(d, N, theta1, theta2, theta3, F, max_iters=1000, return_history=True, verbose=True):
    pbil_start_time = time.time()
    p = np.full(d, 0.5)
    loss_history = np.empty(max_iters)
    pop_min_history = np.empty(max_iters)
    pop_max_history = np.empty(max_iters)
    pop_mean_history = np.empty(max_iters)
    
    for i in range(max_iters):
        iteration_start_time = time.time()
        P = np.random.uniform(size=(N, d)) < p
        f_vals = F(P)
        x = P[np.argmax(f_vals), :] # best individual
        
        loss_history[i] = F(x.reshape(1, -1))
        pop_min_history[i] = np.min(f_vals)
        pop_max_history[i] = np.max(f_vals)
        pop_mean_history[i] = np.mean(f_vals)
             
        p = p * (1 - theta1) + x * theta1
        # changed_flags = np.random.uniform(size=d) < theta2
        # p = p * (1 - changed_flags) + (changed_flags * (p * (1 - theta3) + (np.random.uniform(size=d) < 0.5) * theta3))
        p = p + (-p * theta3 + (np.random.uniform(size=d) < 0.5) * theta3) * (np.random.uniform(size=d) < theta2)
        
        iteration_end_time = time.time()
        time_elapsed = iteration_end_time - iteration_start_time
        if verbose:
            print(f'iteration {i}: loss = {loss_history[i]}')
            print(f'\tmin loss in population = {pop_min_history[i]}')
            print(f'\tmax loss in population = {pop_max_history[i]}')
            print(f'\tmean loss in population = {pop_mean_history[i]}')
            print(f'\ttime of iteration = {time_elapsed}')
    
    pbil_end_time = time.time()
    pbil_time_elapsed = pbil_start_time - pbil_end_time
    if verbose:
        print(f'PBIL total time = {pbil_time_elapsed}')
                  
    if return_history:
        return p, loss_history, pop_min_history, pop_max_history, pop_mean_history
    return p
```

```python
class SpectralData(object):

    def __init__(self, verbose=True):
        self.image_raw_reduced = np.loadtxt('ImageRawReduced.txt')
        self.image_expert_reduced = np.loadtxt('ImageExpertReduced.txt')
        self.classification_rules = np.loadtxt('ClassificationRules.txt')

        if verbose:
            print(f'image_raw_reduced shape: {self.image_raw_reduced.shape}')
            print(f'image_expert_reduced shape: {self.image_expert_reduced.shape}')
            print(f'classification_rules shape: {self.classification_rules.shape}')

class SpectralClassifier(object):

    def __init__(self, data):
        self.data = data

    def classify_instance(self, rules_mask):
        return mode(self.data.classification_rules[rules_mask], axis=0)[0].reshape(-1)

    def classify(self, rules_masks, pool):
        return np.array(pool.map(self.classify_instance, [rules_mask for rules_mask in rules_masks]))

class SpectralLoss(object):

    def __init__(self, data, verbose=True):
        self.data = data
        self.pool = multiprocessing.Pool(processes=multiprocessing.cpu_count())
        if verbose:
            print(f'Number of processes: {multiprocessing.cpu_count()}')

    def loss(self, rules_masks):
        classifier = SpectralClassifier(self.data)
        return np.sum(classifier.classify(rules_masks, self.pool) == self.data.image_expert_reduced, axis=1)

    def close(self):
        self.pool.close()
```

```python
data = np.load('spectral_results_0.01_0.01_0.01_200.npz')
plt.title('Loss change for theta1 = 0.01, theta2 = 0.01, theta3 = 0.01')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.plot(data['loss_history'], label='loss for the best individual')
plt.plot(data['pop_mean_history'], label='mean loss in population')
plt.legend()
```
